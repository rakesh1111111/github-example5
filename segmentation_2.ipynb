```python
import pandas as pd
import numpy as np

df=pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')  
dff=df[0]
dff.drop(dff[ dff["Borough"]=="Not assigned" ].index, axis=0,inplace=True)
dff_gp=dff.groupby("Postal Code").agg(lambda x:','.join(set(x))).rename(columns={'Postal Code': 'PostalCode'})
dff_gp.reset_index(level=0, inplace=True)
dff_gp.loc[dff_gp['Neighborhood']=="Not assigned",'Neighborhood']=dff_gp.loc[dff_gp['Neighborhood']=="Not assigned",'Borough']
dff_gp.columns=['Postalcode','Borough','Neighborhood']
dff_gp
```

```python
dflonglat=pd.read_csv('http://cocl.us/Geospatial_data')
dflonglat.columns=['Postalcode','Latitude','Longitude']
dflonglat
```

```python
final_df=pd.merge(dff_gp, dflonglat, on='Postalcode')
final_df
```

```python
import numpy as np # library to handle data in a vectorized manner

import pandas as pd # library for data analsysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import json # library to handle JSON files

#!conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans

!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library

print('Libraries imported.')
```

```python
df_Toronto=final_df[final_df.Borough.str.contains('Toronto',case=False)]
df_Toronto.reset_index()
df_Toronto
# I decide to use the Borough with the word Toronto
```

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
map_Toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

for lat, lng, borough, neighborhood in zip(df_Toronto['Latitude'], df_Toronto['Longitude'], df_Toronto['Borough'], df_Toronto['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Toronto)  
    
map_Toronto
```

The above is a generalised view of the Toronto Map with the fill ins based on the df_Toronto

```python
Toronto_onehot = pd.get_dummies(df_Toronto[['Borough']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
Toronto_onehot['Neighborhood'] = df_Toronto['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [Toronto_onehot.columns[-1]] + list(Toronto_onehot.columns[:-1])
Toronto_onehot = Toronto_onehot[fixed_columns]

# set number of clusters
kclusters = 4

Toronto_onehot = Toronto_onehot.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(Toronto_onehot)

# check cluster labels generated for each row in the dataframe
kmeans.labels_
```

```python
# add clustering labels
df_Toronto.insert(0, 'Cluster Labels', kmeans.labels_)
df_Toronto
```

```python
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(df_Toronto['Latitude'], df_Toronto['Longitude'], df_Toronto['Neighborhood'], df_Toronto['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
