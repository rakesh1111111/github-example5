```python
import pandas as pd
import numpy as np

df=pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')  
dff=df[0]
dff.drop(dff[ dff["Borough"]=="Not assigned" ].index, axis=0,inplace=True)
dff_gp=dff.groupby("Postal Code").agg(lambda x:','.join(set(x))).rename(columns={'Postal Code': 'PostalCode'})
dff_gp.reset_index(level=0, inplace=True)
dff_gp.loc[dff_gp['Neighborhood']=="Not assigned",'Neighborhood']=dff_gp.loc[dff_gp['Neighborhood']=="Not assigned",'Borough']
dff_gp.columns=['Postalcode','Borough','Neighborhood']
dff_gp
```

```python
dfp=pd.read_html('https://www12.statcan.gc.ca/census-recensement/2016/dp-pd/hlt-fst/pd-pl/Table.cfm?Lang=Eng&T=1201&SR=1&S=22&O=A&RPP=9999&PR=0', skiprows=2)  
df_Pop=dfp[0]
df_Pop.columns=['Postalcode','Population','d1','d2']
df_Population=df_Pop[df_Pop.Postalcode.str.startswith('M')]
df_Population.drop(['d1','d2'],axis=1,inplace=True)
df_Population
```

```python
dflonglat=pd.read_csv('http://cocl.us/Geospatial_data')
dflonglat.columns=['Postalcode','Latitude','Longitude']
dflonglat
```

```python
final_df=pd.merge(dff_gp, dflonglat, on='Postalcode')
final_df=pd.merge(final_df, df_Population, on='Postalcode')
final_df
```

```python
X = np.array(final_df[['Latitude','Longitude']].astype(float))
Y = np.array(final_df['Population'].astype(float))

# set number of clusters
kclusters = 10

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(X,sample_weight = Y)

# check cluster labels generated for each row in the dataframe
kmeans.labels_
kmeans.cluster_centers_
df_cc=pd.DataFrame(kmeans.cluster_centers_,columns=['Latitude_centroid','Longitude_centroid']) 
df_cluster_centroid=df_cc.reset_index()
df_cluster_centroid.columns=['Cluster Labels','Latitude_centroid','Longitude_centroid']
df_cluster_centroid


```

```python
# add clustering labels
df_Toronto.insert(0, 'Cluster Labels', kmeans.labels_)
df_Toronto
```

```python
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(df_Toronto['Latitude'], df_Toronto['Longitude'], df_Toronto['Neighborhood'], df_Toronto['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)

markers_colors = []
for lat, lon, cluster in zip(df_cluster_centroid['Latitude_centroid'], df_cluster_centroid['Longitude_centroid'], df_cluster_centroid['Cluster Labels']):
    folium.CircleMarker(
        [lat, lon],
        radius=11,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.9).add_to(map_clusters)    

map_clusters
```
